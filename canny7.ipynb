```python
import numpy  as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
def show(img):
    if img.ndim==2:
        plt.imshow(img,cmap='gray')
    else:
        plt.imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB)) # 在matplotlib中是RGB，而在opencv中是BGR
    plt.show()
    print(img.ndim)
```

```python
img=cv.imread('picture/7.bmp')
print(type(img))
show(img)
```

```python
gongjian=img[:,:,:]
show(gongjian)
```

```python
gray=cv.cvtColor(gongjian,cv.COLOR_BGR2GRAY)
show(gray)
```

```python
img_edge=cv.Canny(gray,64,128)
show(img_edge)
```

```python
ignore,img_bin=cv.threshold(img_edge,80,255,cv.THRESH_BINARY)
show(img_bin)
print(img_bin.dtype)
```

```python
# 查找轮廓
contours, hierarchy = cv.findContours(img_bin.astype(np.uint8), cv.RETR_EXTERNAL, cv.CHAIN_APPROX_NONE)
# 输出轮廓个数
print(len(contours))
```

```python
gongjian_copy=gongjian.copy()
for i in range(0, len(contours)):
    length = cv.arcLength(contours[i], True)
    print(length)
    # 通过轮廓长度筛选
    if 600>length > 5:
        print(length)
        cv.drawContours(gongjian_copy, contours[i], -1, (0, 0, 255), 2)
```
